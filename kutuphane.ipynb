```python
import tkinter as tk
from tkinter import messagebox
import json

# Dosyadan kitaplarÄ± okuma
def kitaplari_yukle():
    try:
        with open("kitaplar.json", "r", encoding="utf-8") as f:
            return json.load(f)
    except (FileNotFoundError, json.JSONDecodeError):
        return []

# Dosyaya kitaplarÄ± kaydetme
def kitaplari_kaydet():
    with open("kitaplar.json", "w", encoding="utf-8") as f:
        json.dump(kutuphane, f, ensure_ascii=False, indent=4)

# Kitap ekleme fonksiyonu
def kitap_ekle():
    kitap_adi = kitap_adi_entry.get()
    yazar_adi = yazar_adi_entry.get()
    
    if kitap_adi and yazar_adi:
        kutuphane.append({"isim": kitap_adi, "yazar": yazar_adi})
        kitap_listesi.insert(tk.END, f"{kitap_adi} - {yazar_adi}")
        kitap_adi_entry.delete(0, tk.END)
        yazar_adi_entry.delete(0, tk.END)
        kitaplari_kaydet()  # Yeni kitabÄ± dosyaya kaydet
        messagebox.showinfo("BaÅŸarÄ±lÄ±", "Kitap eklendi!")
    else:
        messagebox.showwarning("Hata", "LÃ¼tfen tÃ¼m alanlarÄ± doldurun!")

# Kitap silme fonksiyonu
def kitap_sil():
    secili_kitap = kitap_listesi.curselection()
    if secili_kitap:
        kitap_adi = kitap_listesi.get(secili_kitap)
        kitap_listesi.delete(secili_kitap)
        kutuphane.remove({"isim": kitap_adi.split(" - ")[0], "yazar": kitap_adi.split(" - ")[1]})
        kitaplari_kaydet()  # Kitap silindiÄŸinde dosyayÄ± gÃ¼ncelle
        messagebox.showinfo("BaÅŸarÄ±lÄ±", f"{kitap_adi} silindi!")
    else:
        messagebox.showwarning("Hata", "LÃ¼tfen silmek istediÄŸiniz kitabÄ± seÃ§in.")

# Kitap dÃ¼zenleme fonksiyonu
def kitap_duzenle():
    secili_kitap = kitap_listesi.curselection()
    if secili_kitap:
        eski_kitap = kitap_listesi.get(secili_kitap)
        eski_isim, eski_yazar = eski_kitap.split(" - ")
        
        # KullanÄ±cÄ±dan yeni bilgiler
        yeni_isim = kitap_adi_entry.get()
        yeni_yazar = yazar_adi_entry.get()
        
        if yeni_isim and yeni_yazar:
            kutuphane.remove({"isim": eski_isim, "yazar": eski_yazar})
            kutuphane.append({"isim": yeni_isim, "yazar": yeni_yazar})
            kitap_listesi.delete(secili_kitap)
            kitap_listesi.insert(secili_kitap, f"{yeni_isim} - {yeni_yazar}")
            kitaplari_kaydet()  # GÃ¼ncellenen bilgiyi dosyaya kaydet
            messagebox.showinfo("BaÅŸarÄ±lÄ±", f"{eski_isim} - {eski_yazar} baÅŸarÄ±yla dÃ¼zenlendi!")
        else:
            messagebox.showwarning("Hata", "LÃ¼tfen tÃ¼m alanlarÄ± doldurun.")
    else:
        messagebox.showwarning("Hata", "LÃ¼tfen dÃ¼zenlemek istediÄŸiniz kitabÄ± seÃ§in.")

# Tkinter ana penceresi
pencere = tk.Tk()
pencere.title("KÃ¼tÃ¼phane UygulamasÄ±")
pencere.geometry("600x600")
pencere.configure(bg="#797979")  # Arkaplan rengi (aÃ§Ä±k mavi)

# BaÅŸlÄ±k
baslik_label = tk.Label(pencere, text="ğŸ“š KÃ¼tÃ¼phane UygulamasÄ±", font=("Helvetica", 18, "bold"), bg="#797979",fg="#FFFFFF")
baslik_label.pack(pady=10)

# Kitap Ekleme AlanÄ±
frame_kitap_ekle = tk.Frame(pencere, bg="#e1f5fe")
frame_kitap_ekle.pack(pady=10)

# KitaplarÄ± yÃ¼kleme
kutuphane = kitaplari_yukle()

# Kitap listesi alanÄ±
kitap_listesi_label = tk.Label(pencere, text="ğŸ“– Kitap Listesi",fg="#FFFFFF", bg="#797979", font=("Arial", 16, "bold"))
kitap_listesi_label.place(x=80, y=200)

kitap_listesi = tk.Listbox(pencere,fg="#797979", font=("Arial", 12,"bold"), width=50, height=8)
kitap_listesi.place(x=80, y=240)

# Kitap ekleme alanÄ±
kitap_adi_label = tk.Label(pencere, text="Kitap AdÄ±:",fg="#FFFFFF", bg="#797979", font=("Arial", 14, "bold"))
kitap_adi_label.place(x=80, y=70)

kitap_adi_entry = tk.Entry(pencere, bg="#ffffff",fg="#797979", font=("Arial", 14,), width=30)
kitap_adi_entry.place(x=200, y=70)

yazar_adi_label = tk.Label(pencere, text="Yazar AdÄ±:",fg="#FFFFFF", bg="#797979", font=("Arial", 14, "bold"))
yazar_adi_label.place(x=80, y=110)

yazar_adi_entry = tk.Entry(pencere, bg="#ffffff",fg="#797979", font=("Arial", 14), width=30)
yazar_adi_entry.place(x=200, y=110)

kitap_ekle_buton = tk.Button(pencere, text="Kitap Ekle", bg="#c0c0c0", width =15 , font=("Arial", 12, "bold"), command=kitap_ekle)
kitap_ekle_buton.place(x=200, y=420)

kitap_sil_buton = tk.Button(pencere, text="Kitap Sil",bg="#c0c0c0", width =15 , font=("Arial", 12, "bold"), command=kitap_sil)
kitap_sil_buton.place(x=200, y=460)

kitap_duzenle_buton = tk.Button(pencere, text="Kitap DÃ¼zenle",bg="#c0c0c0", width =15 , font=("Arial", 12, "bold"), command=kitap_duzenle)
kitap_duzenle_buton.place(x=200, y=500)

# Listeyi doldurma
for kitap in kutuphane:
    kitap_listesi.insert(tk.END, f"{kitap['isim']} - {kitap['yazar']}")

# Ana dÃ¶ngÃ¼
pencere.mainloop()
```
